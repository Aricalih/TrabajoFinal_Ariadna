### **Selección de indicadores a utilizar**
**Los indicadores seleccionados para este análisis buscan explorar la relación entre el desarrollo económico, la sostenibilidad ambiental y el acceso a servicios básicos. Específicamente:**

⮕Crecimiento del PIB (anual %): Representa la tasa de crecimiento económico de un país.

⮕Emisiones de CO2 (metric tons per capita): Debido a que refleja el impacto ambiental per cápita, permitiendo comparaciones entre países de diferentes tamaños y niveles de población.

⮕Acceso a electricidad (% de población): Es un indicador de desarrollo y bienestar que refleja la capacidad de un país para proporcionar servicios básicos a sus ciudadanos.

⮕Acceso a electricidad, rural (% de población rural): Proporciona una visión del acceso a servicios básicos en áreas rurales, que suelen ser más desfavorecidas.

⮕Camas de hospital (por 1,000 personas): Sirve como un indicador de la infraestructura de salud de un país, reflejando la capacidad del sistema de salud para atender a la población.

**Estos indicadores fueron seleccionados porque su análisis nos podría proporcionar información sobre cómo el crecimiento económico y el desarrollo de infraestructura afectan o están relacionados con la sostenibilidad ambiental y la calidad de vida.**

```python
seleccion_indicadores = [
    'GDP growth (annual %)', # Crecimiento del PIB (anual %)
    'CO2 emissions (metric tons per capita)', # Emisiones de CO2 per cápita
    'Access to electricity (% of population)', # Acceso a electricidad (% de población)
    'Access to electricity, rural (% of rural population)', # Acceso a electricidad, rural (% de población rural)
    'Hospital beds (per 1,000 people)', # Camas de hospital (por 1,000 personas)
    # Agregar más indicadores si es necesario
]
```

### Filtramos el conjunto de datos por los indicadores seleccionados:

```python
data_filtrada = data_1[data_1['Series Name'].isin(seleccion_indicadores)]
data_filtrada #Visualizamos
```

### Reorganizamos los datos para que cada fila sea un país-año con columnas para cada indicador. Convertimos los años que estaban en columnas en filas, para facilitar el análisis:

```python
mezcla_data = data_filtrada.melt(id_vars=['Country Name', 'Country Code', 'Series Name', 'Series Code'], 
                                 var_name='Year', value_name='Value')

mezcla_data
```

### Convertimos las cadenas 'Year' a enteros y 'Value' a números flotantes, para poder manejar los valores '..' como NaN. Es importante transformar los demás datos a flotantes para realizar cálculos matemáticos como correlaciones:

```python
mezcla_data['Year'] = mezcla_data['Year'].str.extract('(\d+)').astype(int)
mezcla_data['Value'] = pd.to_numeric(mezcla_data['Value'], errors='coerce')

mezcla_data
```

### Ahora aplicamos pivot a los datos para que cada indicador se convierta en una columna:

```python
pivot_data = mezcla_data.pivot_table(index=['Country Name', 'Country Code', 'Year'], 
                                     columns='Series Name', 
                                     values='Value').reset_index()

pivot_data
```

### Limpiamos los datos eliminando filas con valores faltantes en los indicadores de interés:

```python
data_limpia = pivot_data.dropna(subset=seleccion_indicadores)

data_limpia
```

### Realizamos el cálculo de la matriz de correlación:

```python
matriz_correlacion = data_limpia[seleccion_indicadores].corr()
```

### Visualizamos la matriz de correlación, usando seaborn:

```python
plt.figure(figsize=(10, 8))
sns.heatmap(matriz_correlacion, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Matriz de correlación de Indicadores Seleccionados')
plt.show()
```

# *Hallazgos preliminares de la correlación:*

⮕ El **crecimiento del PIB (anual %)** no muestra una correlación fuerte con ninguno de los otros indicadores, ya que todos los valores están cercanos a cero. Esto sugiere que el crecimiento del PIB anual no está directamente relacionado con las emisiones de CO2 per cápita, el acceso a la electricidad, el acceso a la electricidad rural, o la cantidad de camas de hospital por 1,000 personas en los datos analizados.

⮕ Las **emisiones de CO2 (metric tons per capita)** también tienen correlaciones bajas con otros indicadores, excepto con las camas de hospital por 1,000 personas, donde hay una correlación moderada de 0.29. Esto podría sugerir que países con mejor infraestructura de salud tienden a tener mayores emisiones de CO2 per cápita, posiblemente debido a una mayor industrialización y desarrollo económico.

⮕ El **acceso a la electricidad (% de población)** y Acceso a la electricidad, rural (% de población rural) están muy correlacionados entre sí con un coeficiente de 0.96, lo cual es esperado ya que el acceso a la electricidad en áreas rurales contribuye significativamente al acceso general en un país.

⮕ Las **camas de hospital (per 1,000 people)** muestra una correlación moderada con las emisiones de CO2 (0.29) como se mencionó antes, y correlaciones bajas a moderadas con acceso a electricidad, tanto en general (0.41) como en áreas rurales (0.45). Esto puede reflejar que países con mejor infraestructura de salud también tienen mejor infraestructura en servicios básicos como la electricidad.

#  Creación de un gráfico para analizar la dispersión entre el crecimiento del PIB y las emisiones de CO2 per cápita: 

Permite observar la **relación entre dos variables numéricas y cómo una variable se comporta en relación con la otra**. Por ejemplo, al graficar las emisiones de CO2 per cápita contra el crecimiento del PIB, puedes visualizar si hay una tendencia o patrón entre ellas.

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='CO2 emissions (metric tons per capita)', y='GDP growth (annual %)', data=data_limpia)

plt.title('Relación entre Emisiones de CO2 per cápita y Crecimiento del PIB')
plt.xlabel('Emisiones de CO2 (toneladas métricas per cápita)')
plt.ylabel('Crecimiento del PIB (anual %)')
plt.grid(True)
plt.show()

```

# *Hallazgos preliminares del gráfico de dispersión*

El gráfico muestra la relación entre las emisiones de CO2 per cápita y el crecimiento del PIB. En el se puede observar que la mayoría de los puntos están agrupados hacia el extremo inferior izquierdo del gráfico, lo que indica que la mayoría de los datos reflejan niveles bajos a moderados de emisiones de CO2 per cápita y un crecimiento del PIB que varía desde negativo hasta positivo. 

⮕La dispersión horizontal de los puntos a lo largo del eje de crecimiento del PIB sugiere que **no hay una relación lineal clara entre estas dos variables; es decir, el crecimiento del PIB no parece incrementar ni disminuir sistemáticamente con las emisiones de CO2 per cápita**. 

⮕Asimismo, algunos puntos se encuentran lejos del grupo principal, especialmente en la parte superior del gráfico, lo que indica años o países con altos niveles de crecimiento del PIB y también algunos con altas emisiones de CO2 per cápita, pero son relativamente pocos.

Si bien no parece haber una tendencia clara o lineal que indique una relación directa entre las emisiones de CO2 per cápita y el crecimiento del PIB.
